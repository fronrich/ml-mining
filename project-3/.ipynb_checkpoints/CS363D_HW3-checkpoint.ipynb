## Q2 Define the entropy function
Implement a function `calc_entropy` that takes the labels series, s_y, as a parameter. Implement using the definition on p128 in the DM book and only pandas and log2 libraries

```python
def calc_entropy(series):
    # get the ratio
    p = (series.value_counts(normalize = True, sort = False))
    return p.apply ( lambda x: -x*log2(x)).sum()
```

```python
calc_entropy(s_y)
```

```python
def limit_landsize(xmin, xmax):
    df_copy = df_main.copy()
    # limit
    if xmax < xmin:
        df_copy = df_copy[(df_copy['Landsize'] > xmin)]
    else:
        df_copy = df_copy[(df_copy['Landsize'] > xmin) & (df_copy['Landsize'] <= xmax)]
    return df_copy['Type']
```

```python
# entropy in landsize (0, 200]
s_0_200 = limit_landsize(0, 200)
print(s_0_200)
print(calc_entropy(s_0_200))
```

## Q4 Define the Gini Index
Implement the function `calc_gini` that takes the labels series, s_y, as a parameter. Implement using the definition on p128 in the DM book.  Use only pandas library

```python
def calc_gini(series):
    # get the ratio
    p = (series.value_counts(normalize = True, sort = False)).divide(len(series))
    return 1 - p.apply(lambda x: (x * x)).sum()
```

## Q6 Calculate Gain Ratio using unique council area
Calculate the gain ratio (p135) going from the root node, to splitting that node into 29 child nodes based on their `CouncilArea`, where each child node contains a single council area.

```python
df_council_area = df_main['CouncilArea'].value_counts().index.tolist()

sum = 0
length_total = len(df_main['Type'])

for index in df_council_area:
    sector = df_main[df_main['CouncilArea'] == index]
    sector_type = sector['Type']
    ent = calc_entropy(sector_type)
    frac = len(sector) / length_total
    sum += frac * calc_entropy(sector_type)
    
gain_split = calc_entropy(df_main['Type']) - sum
split_info = calc_entropy(df_main['CouncilArea'])
gain_ratio = gain_split / split_info
gain_ratio 
```
