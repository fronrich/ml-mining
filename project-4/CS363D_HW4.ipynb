## Q1.2 Define a function that creates a linear least squares regression model 
This function should take in two parameters, `df_X`, and `s_y` and return the least squares regression model, $\hat{beta}$ (using the notation from the ESLII book chapter 3 and HW2).  You can not use any libraries outside of pandas and numpy. 

````python
def beta_hat(df_X, s_y):
    # define variables
    x = df_X
    y = s_y
    
    #Add intercept column
    intercept =  pd.Series(np.ones(len(df_X)), name="Intercept")
    x = pd.concat([intercept,x],axis=1)
    # convert to matrix
    x = x.to_numpy()
    # x = x.dropna()
    x_t = x.transpose()
    #Calculate beta_hat
    inv = np.linalg.inv(np.matmul(x_t, x))
    temp = np.matmul(inv, x_t)
    b_hat = np.matmul(temp, y)
    
    return x, b_hat, y
````

## Q1.4 Define a function that calculates a regression metric
This function should accept two series of equal length $n$ numpy arrays, `s_y`, and `s_y_hat`. The metric it should calculate is the mean absolute error, $MAE = \sum\limits_{i=1}^n\frac{|{s\_y_i - {s\_y\_hat}_i}|}{n}$ 

Test your function by using the vectors:
```
x = np.array([1,2,3])
y = np.array([2,2,3])
```


````python
def calc_MAE(array_1, array_2):
    length = len(array_1)
    sum = 0
    for x in range (length):
        sum += abs(array_1[x] - array_2[x])
    mae = sum / length
    return mae
````

````python
x = np.array([1,2,3])
y = np.array([2,2,3])
mae = calc_MAE(x,y)
print("MAE: " + str(mae))
````

````python
# create accuracy function
def get_accuracy(series_1, series_2):
    total = len(series_1)
    equal_elements = series_1 == series_2
    num = len(np.where(equal_elements)[0])
    return num / total
````
