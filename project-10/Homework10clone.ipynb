## Part 2 Find which ones have images

```python
root = 'https://collection.blantonmuseum.org/results.html?&layout=objects&format=json&maximumrecords=-1&recordType=objects_1&query=mfs%20all%20'
img_true = 0
img_false = 0
df['Has_Image'] = False
for row, data in df.iterrows():
    if (row == 1372):
        break
    #create url
    art_id = df.iloc[row]['Accession #']
    query_id = str(art_id)
    urlEnd = urllib.parse.quote(query_id)
    url = root + '%22' + urlEnd + '%22'
        
    #get JSON data and get rid of any trailing 
    response  = requests.get(url)
    text = response.text
    text = re.sub(',\s*]',']', text)
    
    #load to JSON
    data = json.loads(text, strict = False)
    
    if (row == 907):
        print(data)
        
    #count images found
    if len(data['objects'][0]['Images']) != 0:
        img_true += 1
        df.at[row,'Has_Image'] = True
```

```python
perct = img_true/(len(df))
print(perct)
print(str(img_true))
```
